```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = pd.read_csv(r"D:\hustle\datasets\Ice_cream selling data.csv")
```

```python
x = data['Temperature (°C)'].values.reshape(-1,1)
y = data['Ice Cream Sales (units)'].values 
```

```python
poly = PolynomialFeatures(degree=3)
x_poly = poly.fit_transform(x)
```

```python
model = LinearRegression()
```

```python
model.fit(x_poly,y)
```

```python
y_pred = model.predict(x_poly)
```

```python
plt.scatter(x, y, color='blue', label='Actual Data')  # Actual data points
plt.plot(x, y_pred, color='red', label='Polynomial Fit')  # Polynomial regression curve
plt.xlabel('Temperature (°C)')
plt.ylabel('Ice Cream Sales (units)')
plt.title('Polynomial Regression: Ice Cream Sales vs Temperature')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error

# Evaluate the model
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

print("R-squared:", r2)
print("Mean Squared Error:", mse)
```
