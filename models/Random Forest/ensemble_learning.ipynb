```python
from sklearn.datasets import load_digits
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
import xgboost as xgb
```

```python
my_data =load_digits()
```

```python
dir(my_data)
```

```python
print(my_data.DESCR)
```

```python
features = pd.DataFrame(my_data.data,columns=my_data.feature_names)
```

```python
features.head()
```

```python
target = pd.DataFrame(my_data.target,columns=['Target'])
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target,train_size=0.7)
```

```python
pca = PCA(n_components=0.7)
# n_components in % 0-1  0.6
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
pca.fit(x_train)
pca.fit(x_test)
```

```python
x_train_transformed = pca.transform(x_train)
x_test_transformed =  pca.transform(x_test)
```

```python
x_train_transformed.shape
```

```python
pd.DataFrame(x_train_transformed)
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(x_train_transformed,y_train)
```

```python
log_model.score(x_test_transformed,y_test)
```

```python
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
```

```python
bg_class = BaggingClassifier(estimator=LogisticRegression(),n_estimators=10,bootstrap=True)
#n_estmator is samples, here 10 samples
```

```python
bg_class.fit(x_train_transformed,y_train)
```

```python
y_pred = bg_class.predict(x_test_transformed)
```

```python
accuracy_score(y_pred,y_test)
```

## RandomForest

```python
random_forest = RandomForestClassifier(n_estimators=100)
```

```python
random_forest.fit(x_train_transformed,y_train)
```

```python
y_pred = random_forest.predict(x_test_transformed)
```

```python
accuracy_score(y_pred,y_test)
```
