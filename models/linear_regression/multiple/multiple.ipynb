```python
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D scatter plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the actual data
ax.scatter(df['age'], df['experience'], df['income'], color='blue', label='Actual Data')

# Predicted data
ax.plot(df['age'], df['experience'], reg.predict(df[['age', 'experience']]), color='red', label='Regression Line')

# Set labels
ax.set_xlabel('Age')
ax.set_ylabel('Experience')
ax.set_zlabel('Income')
plt.title('3D Plot of Age, Experience, and Income')
plt.legend()
plt.show()
```
